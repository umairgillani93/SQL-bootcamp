## SQL Joins:

-------------- INNER JOIN -----------

select count(*) from payment
inner join customer
on payment.customer_id = customer.customer_id;

-- if the column we are choosing is common in both the tables 
-- we may want to give the table name from which we want to get it
-- like here, customer_id is common in both
-- so we pick it this way payment.customer_id
select payment_id, payment.customer_id from payment
inner join customer
on payment.customer_id = customer.customer_id;


select staff_id, payment_id from payment
inner join customer
on payment.customer_id = customer.customer_id

select payment_id, payment.customer_id, first_name, email, store_id
from payment
inner join customer
on payment.customer_id = customer.customer_id


SELECT payment_id from customer
inner join payment
on customer.customer_id = payment.customer_id
---------------------------------------------------------------------
-------------- FULL OUTER JOIN ---------------------

-- Full outer join grabs both all the rows in both the tables 

select * from payment
full outer join customer
on payment.customer_id = customer.customer_id


select * from payment
full outer join customer
on payment.customer_id = customer.customer_id 
where payment.payment_id is null or customer.customer_id is null;

----------- LEFT OUTER JOIN -------------
-- LEFT OUTER JOIN grabs all the rows of table A and matching rows of table --

select * from payment
left outer join customer
on payment.customer_id = customer.customer_id

select * from payment
left join customer
on payment.customer_id = customer.customer_id
where payment.customer_id is null

select film.film_id, title, inventory_id, store_id
from film
left join inventory
on inventory.film_id = film.film_id
where inventory.film_id is null
